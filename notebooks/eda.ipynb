```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
train = pd.read_csv('D:/kifiya AI/credit-risk-model/data/raw/data.csv', parse_dates=['TransactionStartTime'])
print(train.shape)
```

missing

```python
ms = train.isnull().mean().sort_values(ascending=False)
ms[ms>0]
```

numeric dist

```python
train[['Amount','Value']].hist(bins=50, figsize=(8,3))
```

categorical

```python
train['ChannelId'].value_counts().head(10).plot.barh()
```

fraud ratio by channel

```python
(train.groupby('ChannelId')['FraudResult'].mean().sort_values(ascending=False)
 .plot.barh(title='Fraud rate'))
```

correlation

```python
sns.heatmap(train[['Amount','Value','FraudResult']].corr(), annot=True)
```

outliers

```python
q1,q3 = train['Amount'].quantile([.25,.75])
iqr = q3-q1
print('Outliers>', q3+1.5*iqr, (train['Amount'] > q3+1.5*iqr).sum())
```
